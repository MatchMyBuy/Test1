```python
from google.colab import files
from google.oauth2 import service_account
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload

# ðŸ”¹ Step 1: Upload your files
# ðŸ‘‰ Choose BOTH: your JSON key + one audio file (.mp3 or .opus)
uploaded = files.upload()

# ðŸ”¹ Step 2: Setup credentials
SERVICE_ACCOUNT_FILE = [f for f in uploaded.keys() if f.endswith('.json')][0]
SCOPES = ['https://www.googleapis.com/auth/drive.file']
FOLDER_ID = '11IO1jT4buWeh5xTqw0CIJNEUU030w4uA'  # Your Drive folder ID

credentials = service_account.Credentials.from_service_account_file(
    SERVICE_ACCOUNT_FILE, scopes=SCOPES)
drive_service = build('drive', 'v3', credentials=credentials)

# ðŸ”¹ Step 3: Find the audio file
audio_file = [f for f in uploaded.keys() if f.endswith('.mp3') or f.endswith('.opus')][0]

# ðŸ”¹ Step 4: Upload to Google Drive
media = MediaFileUpload(audio_file, mimetype='audio/mpeg')  # 'audio/mpeg' works for both .mp3 and .opus
file_metadata = {'name': audio_file, 'parents': [FOLDER_ID]}

uploaded_file = drive_service.files().create(
    body=file_metadata,
    media_body=media,
    fields='id, name'
).execute()

print("âœ… Uploaded to Drive:", uploaded_file['name'])
```
